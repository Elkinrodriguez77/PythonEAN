## Gráfica de línea

Un gráfico de línea es utilizado para visualizar la información de cada variable para un individuo en específico. Es muy útil para representar series de tiempo.\

A continuación se realiza un gráfico de la venta realizada en cada uno de los registros. Observe que en el eje x se coloca el número de registro y en el eje y la respectiva venta.

```python
# Asegurarse de que los datos estén en el formato correcto
df['Valor_Venta'] = df['Valor_Venta'].astype(float)

plt.figure(figsize=(10,6))
plt.plot(df['Valor_Venta'], marker='o', linestyle='-')
plt.title('Venta realizada en cada registro')
plt.xlabel('Número de Registro')
plt.ylabel('Venta')
plt.xticks(ticks=range(len(df)), labels=range(1, len(df)+1))
plt.grid(True)
plt.show()
```

## Manipulación de gráficos

Varios objetos del entorno de la gráfica pueden ser manipulados.

```python
# Gráfica
plt.plot(df['Valor_Venta'], marker='o', linestyle='-', label='Ventas')

# Cuadrícula
plt.grid(True)

# Colocar una línea vertical en 5
plt.axvline(x=4, color='r', linestyle='--')

# Línea horizontal que termina en la mitad de la gráfica
plt.axhline(y=df['Valor_Venta'].max()/2, color='g', linestyle='--')

# Banda entre 2 y 3 con una transparencia del 30% de color azul
plt.axvspan(2, 3, facecolor='b', alpha=0.3)

# Colocar texto dentro del gráfico
plt.text(1, df['Valor_Venta'].max() * 0.9, 'Texto dentro del gráfico', fontsize=12)

# Nombre eje x
plt.xlabel('Número de Registro')

# Nombre eje y
plt.ylabel('Venta')

# Título del gráfico
plt.title('Venta realizada en cada registro')

# Leyenda
plt.legend()

# Etiquetas para el eje x comenzando desde 1
plt.xticks(ticks=range(len(df)), labels=range(1, len(df)+1))

# Visualización
plt.show()
```

## Gráfica de descuento y comisión

Varias variables en un mismo gráfico.

```python
plt.figure(figsize=(10,6))

# Gráfica de Descuento
plt.plot(df['Descuento'], marker='o', linestyle='-', label='Descuento')

# Gráfica de Comisión
plt.plot(df['comision'], marker='x', linestyle='-', label='Comisión')

# Gráfica de Descuento - Comisión
plt.plot(df['Descuento']-df['comision'], marker='x', linestyle='-', label='Descuento - Comisión')

# Nombre eje x
plt.xlabel('Número de Registro')

# Nombre eje y
plt.ylabel('Valor')

# Título del gráfico
plt.title('Descuento y Comisión en cada registro')

# Leyenda
plt.legend()

# Etiquetas para el eje x comenzando desde 1
plt.xticks(ticks=range(len(df)), labels=range(1, len(df)+1))

# Visualización
plt.show()

```

## Varios Gráficos

Es posible realizar varios gráficos en una misma ventana utilizando el comando $plt.subplot$

```python
# Divide la ventana en varias gráficos (3 filas, 1 columna)
fig, axs = plt.subplots(3, 1, figsize=(7,10))

# Primer gráfico: Gráfico de ventas
axs[0].plot(df['Valor_Venta'], marker='o', linestyle='-')
axs[0].set_xlabel('Número de Registro') # Nombre eje x
axs[0].set_ylabel('Venta') # Nombre eje y
axs[0].set_title('Venta en cada registro') # Título del gráfico
axs[0].set_xticks(range(len(df))) # Etiquetas para el eje x comenzando desde 0
axs[0].set_xticklabels(range(1, len(df)+1)) # Etiquetas para el eje x comenzando desde 1

# Segundo gráfico: Gráfico de descuento
axs[1].plot(df['Descuento'], marker='o', linestyle='-', color='g')
axs[1].set_xlabel('Número de Registro') # Nombre eje x
axs[1].set_title('Descuento en cada registro') # Título del gráfico
axs[1].set_xticks(range(len(df))) # Etiquetas para el eje x comenzando desde 0
axs[1].set_xticklabels(range(1, len(df)+1)) # Etiquetas para el eje x comenzando desde 1

# Tercer gráfico: Gráfico de comisión
axs[2].plot(df['comision'], marker='o', linestyle='-', color='r')
axs[2].set_xlabel('Número de Registro') # Nombre eje x
axs[2].set_title('Comisión en cada registro') # Título del gráfico
axs[2].set_xticks(range(len(df))) # Etiquetas para el eje x comenzando desde 0
axs[2].set_xticklabels(range(1, len(df)+1)) # Etiquetas para el eje x comenzando desde 1

plt.tight_layout() # Ajusta el diseño para evitar superposición
plt.show() # Visualización
```

## Diagramas de dispersión

Un diagrama de dispersión 2D es un gráfico de puntos que representa el valor que toma cada individuo en dos variables distintas. Se grafica el de las variables Venta y Descuento.

```python
#Gráfico de Ventas vs Descuento
plt.scatter(df['Valor_Venta'],df['Descuento'])
#Nombre eje x
plt.xlabel('Ventas')
#Nombre eje y 
plt.ylabel('Descuento')
#Titulo del gráfico
plt.title('Ventas vs Descuento')
```

También, es posible cambiar el tipo de marca que se quiere visualizar con el comando $marker$, el tamaño con $s$ y el color con $c$. A continuación, se crea un vector aleatorio $x$ para que cada marca quede con un color distinto.

```python
plt.figure(figsize=(10,6))

# Gráfico de dispersión de la Comisión vs el Descuento
plt.scatter(df['Descuento'], df['comision'], c='b', marker='o')

# Nombre eje x
plt.xlabel('Descuento')

# Nombre eje y
plt.ylabel('Comisión')

# Título del gráfico
plt.title('Gráfico de dispersión de la Comisión vs el Descuento')

plt.show()
```

Ahora, un gráfico con todas las posibles diagramas de dispersión.

```python
# Gráfico con todas las posibles diagramas de dispersión
sns.pairplot(df, diag_kind='kde', markers='o')

plt.show()
```
