## Calcule las siguientes funciones del número 15.6:

* Seno
* Coseno
* Su cuadrado
* Su raíz cuadrada
* Logaritmo
* Halle el mayor menor entero del número
* Halle el menor mayor entero del número
* Halle el valor absoluto del negativo del número
* El factorial del número

```python
import math as m

numero = 15.6
num_absoluto = abs(numero)
num_neg = -numero

# Seno
seno = m.sin(numero)

# Coseno
coseno = m.cos(numero)

# Cuadrado
num_cuadrado = numero**2

# Raiz cuadrada modo 1
raiz_cuadrada = numero**(1/2)

# Raiz cuadrada modo 2
raiz_cuadrada2 = m.sqrt(numero)

# Logaritmo
logaritmo = m.log(numero)

# Mayor entero del numero
mayor_entero = m.ceil(num_absoluto)

# Menor entero del numero
menor_entero = m.floor(num_absoluto)

# Valor absoluto del negativo
valor_absoluto = abs(num_neg)

# Factorial del número (tomando la parte entera):
entero = int(num_absoluto)
factorial = m.factorial(entero)

print("Seno:", seno)
print("Coseno:", coseno)
print("Cuadrado:", num_cuadrado)
print("Raiz cuadrada:", raiz_cuadrada)
print("Raiz cuadrada2:", raiz_cuadrada2)
print("Logaritmo:", logaritmo)
print("Mayor entero:", mayor_entero)
print("Menor entero:", menor_entero)
print("Valor absoluto:", valor_absoluto)
print("Factorial:", factorial)
```

## Dados los números 15 y 12 calcule:

* Su suma
* Su multiplicación
* El primer número elevado al segundo
* La raíz del primero aplicada al segundo
* El cociente entre dividir el primero entre el segundo
* El residuo de dividir el primero entre el segundo
* El máximo común divisor de ambos números

```python
numeros = [15,12]

# Suma
suma = numeros[0] + numeros[1]

print(f"La suma entre {numeros[0]} y {numeros[1]} es {suma}")

# Multiplicación
suma = numeros[0] * numeros[1]

print(f"La multiplicación entre {numeros[0]} y {numeros[1]} es {suma}")

# Elevado
elevado = numeros[0]**numeros[1]

print(f"El primer numero ({numeros[0]}) elevado el segundo ({numeros[1]}) es {elevado}")

# Residuo
residuo = numeros[0]%numeros[1]

print(f"El residuo es {residuo}")

# Cociente
cociente = numeros[0] / numeros[1]

print(f"El cociente entre los numeros {numeros[0]} y {numeros[1]} es {cociente}")

# Maximo común divisor:
max_divisor = m.gcd(numeros[0], numeros[1])

print(f"El maximo común divisor de los numeros es {max_divisor}")
```

## A partir del vector (1, 2, 3, 5, 8, 13, 21, 44, 65) defina los siguientes vectores:

* (3, 6, 9, 15, 24, 39, 63, 132, 195)
* (1, 8, 27, 125, .., 274625)
* Extraiga los componentes impares del vector.
* Un vector conformado por la suma de dos componentes consecutivos. Este vector
* Tiene 8 elementos.
* Realice tres gr´aficos con los anteriores vectores

```python
import matplotlib.pyplot as plt

# Vector original
vector = [1,2,3,5,8,13,21,44,65]

# Primer vector
vector1 = [3 * valor for valor in vector]
print(f"El vector resultante al ser multiplicado por 3, es igual a {vector1}")

# Segundo vector
vector2 = [valor**3 for valor in vector]
print(f"El vector resultante elevado al cubo, es igual a {vector2}")

# Tercer vector
vector3 = [valor for valor in vector if valor % 2 == 1]
print(f"El vector resultante únicamente con valores impares, es igual a {vector3}")

# Cuarto vector
vector4 = [vector[i] + vector[i+1] for i in range(len(vector)-1)]
print(f"El vector resultante al sumar dos componentes consecutivos, es igual a {vector4}")

# Gráficos
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(vector1, 'o-')
plt.title('Vector1: Multiplicado por 3')

plt.subplot(2, 2, 2)
plt.plot(vector2, 'o-')
plt.title('Vector2: Al cubo')

plt.subplot(2, 2, 3)
plt.plot(vector3, 'o-')
plt.title('Vector3: Componentes impares')

plt.subplot(2, 2, 4)
plt.plot(vector4, 'o-')
plt.title('Vector4: Suma de componentes consecutivos')

plt.tight_layout()
plt.show()
```
